# CSGO Feature Engineering

This notebook inspects the engineered aim-telemetry features used by the adapted NoScope-Bio pipeline.

```python
from pathlib import Path
import sys
import pandas as pd

ROOT = Path('..').resolve()
sys.path.insert(0, str(ROOT / 'src'))

from noscope_bio.csgo_pipeline import engineer_csgo_engagement, load_csgo_archive_arrays, CSGO_FEATURE_COLUMNS
```

```python
legit, cheaters = load_csgo_archive_arrays()
legit_frame = engineer_csgo_engagement(legit[0, 0], 'L00000_E00', 'L00000', 'legit', 'example', 0)
cheat_frame = engineer_csgo_engagement(cheaters[0, 0], 'C00000_E00', 'C00000', 'cheater', 'example', 0)
```

```python
cols = [
    't', 'view_yaw', 'view_pitch', 'target_error', 'angular_speed',
    'stability_score', 'on_target_tight', 'fire_input', 'flick_event',
    'fire_motion_coupling'
]
legit_frame[cols].head(20)
```

```python
cheat_frame[cols].head(20)
```

```python
summary = pd.DataFrame({
    'feature': CSGO_FEATURE_COLUMNS,
    'legit_mean': [float(legit_frame[c].mean()) for c in CSGO_FEATURE_COLUMNS],
    'cheat_mean': [float(cheat_frame[c].mean()) for c in CSGO_FEATURE_COLUMNS],
})
summary.head(20)
```
