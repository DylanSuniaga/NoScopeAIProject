# CSGO Archive Overview

Quick exploration notebook for the CSGO aim archive under `real_data/archive`.

```python
from pathlib import Path
import sys
import numpy as np
import pandas as pd

ROOT = Path('..').resolve()
sys.path.insert(0, str(ROOT / 'src'))

from noscope_bio.csgo_pipeline import RAW_FEATURE_NAMES, load_csgo_archive_arrays
```

```python
legit, cheaters = load_csgo_archive_arrays()
print('legit', legit.shape, legit.dtype)
print('cheaters', cheaters.shape, cheaters.dtype)
```

```python
rows = []
for label, arr in [('legit', legit), ('cheater', cheaters)]:
    for idx, feature in enumerate(RAW_FEATURE_NAMES):
        values = arr[..., idx]
        rows.append({
            'label': label,
            'feature': feature,
            'mean': float(values.mean()),
            'std': float(values.std()),
            'min': float(values.min()),
            'max': float(values.max()),
        })
pd.DataFrame(rows)
```
